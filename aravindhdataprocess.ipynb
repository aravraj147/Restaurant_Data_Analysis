### **TASK3- GEOSPATIAL ANALYSIS**

```python
coordinates=list(new_df[['Latitude','Longitude']].to_records(False))
len(coordinates)
```

```python
from folium.map import Html
import folium
from folium.plugins import MarkerCluster
from IPython.display import display
from IPython.display import HTML

def create_map(coords_subset):
    map_center = coords_subset[0]
    my_map = folium.Map(location=map_center, zoom_start=20)

    marker = MarkerCluster().add_to(my_map)

    for coord in coords_subset:
        lat, lon = coord
        if pd.notna(lat) and pd.notna(lon):
          folium.Marker(location=[lat, lon],).add_to(marker)

    return my_map

chunk_size = len(coordinates) // 4
chunks = [coordinates[i:i + chunk_size] for i in range(0, len(coordinates), chunk_size)]

print('Visualization of Restaurants as Clusters:')
for i, chunk in enumerate(chunks):
    map_clusters = create_map(chunk)
    display('\n',map_clusters)
```

```python
from scipy.stats import f_oneway

groups = [group['Aggregate rating'] for name, group in new_df.groupby('City')]
f_stat, p_val = f_oneway(*groups)

# Print the results
print("F-Statistic:", f_stat)
print("P value:", p_val)

print('Correlation of Restaurant ratings among all cities:\n')
if p_val < 0.05:
    print("There's a significant difference in ratings based on location.")
else:
    print("There's no significant difference in ratings based on location.")
```
